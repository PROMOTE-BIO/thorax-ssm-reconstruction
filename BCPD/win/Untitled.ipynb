```python
import numpy as np
```

```python
corr = np.loadtxt("output_e.txt", skiprows=1, usecols=(0,1))

target = np.loadtxt("bunny-x.txt", delimiter=",")
source= np.loadtxt("bunny-y.txt", delimiter=",")
deformada= np.loadtxt("output_y.txt")
```

```python
def constroi_matriz(truth,deformed,cor):
    X=[] #pontos do target/ground-truth
    B=[] #pontos da deformada
    for i in range(len(cor)):
        index=int(cor[i][1])
        X+=[truth[i]]
        B+=[deformed[index-1]]
    return X,B
```

```python
def menos(matrizA,matrizB):
    resultado=[]
    for i in range(len(matrizA)):
        resultado+=[matrizA[i]-matrizB[i]]
    return np.array(resultado)

def r(X,B,cor):
    X_B=menos(X, B)
    numerador=np.trace(np.outer(X_B,X_B))
    denominador=len(cor)
    return (numerador/denominador)**(1/2)

def accuracy(source, target, deformada, corr):
    X,B=constroi_matriz(target, deformada, corr)
    fracao=r(source,B,corr)/r(source,X,corr)
    return 1-fracao
```

```python
valor=accuracy(source, target, deformada, corr)
print(valor)
```
